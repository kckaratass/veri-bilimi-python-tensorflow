# groupby

```python
maasSozlugu = { "Departman" : ["Yazılım","Yazılım","Pazarlama","Pazarlama","Hukuk","Hukuk"],
                "Calisan Ismi": ["Ahmet", "Mehmet","Kadir", "Burak", "Zeynep", "Fatma" ],
                "Maas" : [100,150,200,300,400,500]
               }
```

```python
maasDataFrame = pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
groupObjesi = maasDataFrame.groupby("Departman")
```

```python
groupObjesi.count() 
```

```python
groupObjesi.mean(numeric_only=True)
```

```python
groupObjesi.min()
```

```python
groupObjesi.max()
```

```python
groupObjesi.describe()
```
